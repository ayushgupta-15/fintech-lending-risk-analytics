# 01 Data Cleaning
This notebook cleans the dataset based on the established business rules.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/loans_full_schema.csv')

# Create default_flag and outcome_eligible
df['default_flag'] = df['loan_status'].isin(['Charged Off', 'Default']).astype(int)
df['outcome_eligible'] = df['loan_status'].isin(['Charged Off', 'Default', 'Fully Paid']).astype(int)

# Handle Missing Values
df['months_since_last_delinq'] = df['months_since_last_delinq'].fillna(999)
df['months_since_90d_late'] = df['months_since_90d_late'].fillna(999)
df['months_since_last_credit_inquiry'] = df['months_since_last_credit_inquiry'].fillna(999)
df['num_accounts_120d_past_due'] = df['num_accounts_120d_past_due'].fillna(0)
df['emp_title'] = df['emp_title'].fillna('Unknown')
df['emp_length'] = df['emp_length'].fillna(-1)

# Create derived risk segments
def get_dti_band(dti):
    if pd.isna(dti): return 'Unknown'
    if dti < 10: return '< 10%'
    if dti < 20: return '10-20%'
    if dti < 30: return '20-30%'
    if dti < 40: return '30-40%'
    return '> 40%'

df['dti_band'] = df['debt_to_income'].apply(get_dti_band)
df['risk_segment'] = df['grade'] + ' - ' + df['dti_band']

# Save cleaned data
df.to_csv('../data/processed/cleaned_loans.csv', index=False)
print('Data cleaned and saved to data/processed/cleaned_loans.csv')
```
