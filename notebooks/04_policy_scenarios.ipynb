# 04 Policy Scenarios
Simulates Baseline, Balanced, and Conservative policies retroactively to see impact on volume and risk.

```python
import pandas as pd

df = pd.read_csv('../data/processed/cleaned_loans.csv')
eligible = df[df['outcome_eligible'] == 1]

def evaluate_policy(name, data):
    vol = len(data)
    funded = data['loan_amount'].sum()
    defaults = data['default_flag'].sum()
    default_rate = defaults / vol if vol > 0 else 0
    exposure = data[data['default_flag']==1]['loan_amount'].sum()
    return {'Policy': name, 'Volume': vol, 'Funded ($)': funded, 'Default Rate': default_rate, 'Default Exposure ($)': exposure}

baseline = eligible.copy()
balanced = eligible[(~eligible['grade'].isin(['G'])) & (eligible['debt_to_income'] <= 35)]
conservative = eligible[(~eligible['grade'].isin(['E', 'F', 'G'])) & (eligible['debt_to_income'] <= 25)]

results = pd.DataFrame([
    evaluate_policy('Baseline', baseline),
    evaluate_policy('Balanced', balanced),
    evaluate_policy('Conservative', conservative)
])

print(results.to_markdown(index=False))
```
