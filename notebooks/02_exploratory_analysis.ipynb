# 02 Exploratory Data Analysis
This notebook explores the distributions of key risk variables (Grade and DTI) to determine defensible thresholds for policy scenarios.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/cleaned_loans.csv')
eligible = df[df['outcome_eligible'] == 1]

print(f'Total eligible loans: {len(eligible)}')
print(f'Total defaults: {eligible.default_flag.sum()}')
```

```python
# DTI Distribution
plt.figure(figsize=(10,5))
sns.histplot(eligible['debt_to_income'], bins=30, kde=True)
plt.title('Distribution of Debt-to-Income (DTI)')
plt.show()
```

```python
# Default Rate by Grade
grade_stats = eligible.groupby('grade')['default_flag'].agg(['count', 'mean'])
grade_stats.columns = ['Total Loans', 'Default Rate']
print(grade_stats)
```
